```python
from typing import Optional

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BTS:
    def __init__(self):
        self.__root = None

    def insert(self, value):
        # Create new Node
        node = Node(value)

        # Chaeck if exists root
        if not self.__root:
            print("The tree is empty, adding node as root!")
            self.__root = node
            return True
        else:
            self.__insert(self.__root, node)
            # if __insert[1]:
            #     self.__level += 1
            # __insert

    def search(self, value):
        if not self.__root:
            print("Tree is empty!")
            return False

        return self.__search(self.__root, value)

    def remove(self, value):
        if not self.__root:
            print("Tree is empty!")
            return False

        return self.__remove(value)

    def pre_order(self, output = []):
        if not self.__root:
            print("The tree is empty!")
        else:
            self.__pre_order(self.__root, output)
            return output

    def pos_order(self, output = []):
        if not self.__root:
            print("The tree is empty!")
        else:
            self.__pos_order(self.__root, output)
            return output

    def in_order(self, output = []):
        if not self.__root:
            print("The tree is empty!")
        else:
            self.__in_order(self.__root, output)
            return output

    def __insert(self, parent, node):
        # Can't have values duplicated
        if parent.value == node.value:
            print("The node already exists in this tree!")
            return False

        if node.value < parent.value:
            # Set the actual node to left side (If not exists)
            if not parent.left:
                parent.left = node
                return True

            # Pass the "problem" to left side of the tree
            self.__insert(parent.left, node)
        else:
            if not parent.right:
                parent.right = node
                return True

            # Pass the "problem" to the left side of the tree
            self.__insert(parent.right, node)

    def __pre_order(self, parent, output:list):
        if parent:
            output.append(parent.value)
            self.__pre_order(parent.left, output)
            self.__pre_order(parent.right, output)

    def __pos_order(self, parent, output):
        if parent:
            self.__pos_order(parent.left, output)
            self.__pos_order(parent.right, output)
            output.append(parent.value)

    def __in_order(self, parent, output):
        if parent:
            self.__in_order(parent.left, output)
            output.append(parent.value)
            self.__in_order(parent.right, output)

    def __search(self, node: Node, value:int) -> Optional[Node]:
        # Only when have road to node
        if node and node.value:
            if node.value == value:
                return node
            elif value > node.value:
                # Send search to right side
                return self.__search(node.right, value)
            else:
                # Send search to left side
                return self.__search(node.left, value)

        # Tree not have more node and value not find!
        # Return None by default

    def __remove(self, value):
        node_find = self.__search(self.__root, value)
        if node_find and not node_find.left and not node_find.right:
            print(f'The node "{node_find.value}" has been removed!')
            # I try remove refernce, but not funcs
            node_find.value = None
            return True
        else:
            print(f'The node "{node_find.value}" have sub-trees')
            return False

##########################################
# Tests

bts1 = BTS()
int_values = [45, 12, 23, 34, 56, 32, 45, 14, 78, 98, 54, 43, 90]

for val in int_values:
    bts1.insert(val)

# Roads
pre = bts1.pre_order()
in_o = bts1.in_order()
pos = bts1.pos_order()

# Search
search1 = bts1.search(45)
search2 = bts1.search(90)
search3 = bts1.search(78)
search4 = bts1.search(32)

print("\n------------Init------------")
print(f'{pre=}')
print(f'{in_o=}')
print(f'{pos=}')
print("------------Search Before Remove------------")
try:
    print(f'{search1.value=}')
except Exception as e:
    print("Error search1")

try:
    print(f'{search2.value=}')
except Exception as e:
    print("Error search2")

try:
    print(f'{search3.value=}')
except Exception as e:
    print("Error search3")

try:
    print(f'{search4.value=}')
except Exception as e:
    print("Error search4")

print("------------Remove------------")
#
# Try remove
rem1 = bts1.remove(45)
rem2 = bts1.remove(90)
rem3 = bts1.remove(78)
rem4 = bts1.remove(32)

print(f'{rem1=}')
print(f'{rem2=}')
print(f'{rem3=}')
print(f'{rem4=}')
print("------------Search After Remove------------")
# Try search removed nodes
search5 = bts1.search(45)
search6 = bts1.search(90)
search7 = bts1.search(78)
search8 = bts1.search(32)

try:
    print(f'{search5.value=}')
except Exception as e:
    print("Error search5")

try:
    print(f'{search6.value=}')
except Exception as e:
    print("Error search6")

try:
    print(f'{search7.value=}')
except Exception as e:
    print("Error search7")

try:
    print(f'{search8.value=}')
except Exception as e:
    print("Error search8")
print("\n------------End------------")
```
